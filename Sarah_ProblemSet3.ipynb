### 2. Read the data set to your notebook. Modify the dataset to replace \N values everywhere in the dataset with NaN. Capitalize the first letter of each column name and assign the new names to the columns of the dataset.

```python
df = pd.read_csv('master_data_ready.csv')
```

```python
# Replace \N value with NaN everywhere.
import numpy as np
df = df.replace(r"\N", np.nan)
```

```python
df.sample(1)
```

```python
# Caplitalize the first letter of each column name and assign new names to the columns of the dataset.
df.columns = [x.capitalize() for x in df.columns]
```

```python
df.sample(2)
```

### 10. Using Master_Clean: Use the contains method to display city, state, county and regions of jobs that have “New York” or “Los Angeles-Long” in their region state value.

### 11. Using Master_Clean: Display company, vertical, posting date and salaries for all jobs posted during the years 2017 and 2018 in “New York” or “Seattle” region states that had salaries higher than 80,000. Exclude NaN from company column.

```python
Master_Clean.loc[((Master_Clean['Post_date'].str.contains('2017')) 
                 | (Master_Clean['Post_date'].str.contains('2018'))) 
                 & ((Master_Clean['Region_state'].str.contains('New York', na=False)) 
                 | (Master_Clean['Region_state'].str.contains('Seattle', na=False))) 
                 & (Master_Clean['Salary'] > 80000)][['Company', 'Vertical', 'Post_date', 'Salary']].dropna(subset=['Company'])
```

### 12. Using Master_Clean: Calculate the percentage of jobs offering a six-figure salary or higher in the state “NY” in 2018.

```python
six_fig_NY_2018 = Master_Clean.loc[(Master_Clean['Salary'] >= 100000) & (Master_Clean['State'] == 'NY') 
                 & (Master_Clean['Post_date'].str.contains('2018'))]

all_NY_2018 = Master_Clean.loc[(Master_Clean['State'] == 'NY') & (Master_Clean['Post_date'].str.contains('2018'))]
```

```python
(len(six_fig_NY_2018) / len(all_NY_2018)) * 100
```

Jobs offering a sex-figures salary or higher in NY in 2018 was 37% of all jobs in NY in 2018.

### 13. Using Master_Clean: Calculate the percentage of job postings for the company “Wells Fargo” where the salary offered is above the mean of all salaries in the states of “NY” or “CA”.

```python
# The mean of all salaries in the states of 'NY' or 'CA'
mean_sal_nyca = Master_Clean.loc[(Master_Clean['State'] == 'NY') 
                                 | (Master_Clean['State'] == 'CA')].agg({'Salary':'mean'}).item()
mean_sal_nyca
```

```python
# The percentage of job postings for the company "Wells Fargo", of all jobs
jobs_wf_nyca = Master_Clean.loc[(Master_Clean['Company'] == 'Wells Fargo') & (Master_Clean['Salary'] > mean_sal_nyca)]
jobs_wf = Master_Clean.loc[Master_Clean['Company'] == 'Wells Fargo']
```

```python
(len(jobs_wf_nyca) / len(jobs_wf)) * 100
```

The percentage of Wells Fargo job postings that are above the mean of all salaries in NY or CA, is about 74% of all Wells Fargo job postings.
