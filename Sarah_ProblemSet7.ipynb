### 1. Regression Task. You are a data scientist with an e-commerce firm. The firm wants you to analyze their customer data and predict the yearly amount spent.
    a) Explore correlations between customer features and the variable you are trying to predict. Show your exploration by displaying scatter charts or other visualization techniques you used.
    b) Use numerical features of the customers as your customer features X and variable y (to predict) as the “Yearly Amount Spent” column.

```python
df = pd.read_csv("Ecommerce_data.csv")
```

```python
# a) Explore correlations between customer features and the variable you are trying to predict. 
# Show your exploration by displaying scatter charts or other visualization techniques you used.
corr = df.corr()
corr
```

```python
fig, ax = plt.subplots(figsize=(9,6))

sns.heatmap(corr, ax=ax)
ax.set_title("Correlation between Customer Features and the Variables", fontsize=20);
```

```python
# b) Use numerical features of the customers as your customer features X and variable y 
# (to predict) as the “Yearly Amount Spent” column.

results = smf.ols('Q("Yearly Amount Spent") ~ Q("Length of Membership")', df).fit()
```

```python
print(results.summary())
```

```python
df['Prediction'] = results.predict()
```

## 2. Given customer characteristics like age, income, education, payment history, etc. your task is to predict whether they will default or not (Choose any variable or combination of variables from the dataset that you think may work best to predict the likeliness of default).
    a) Apply logictic regression to perform classification task. Use suitable method to evaluate the performance of your approach.
    b) Apply K-nearest neighbors to perform classification. Use a suitable method to evaluate the performance of your approach. Fine the most suitable value of ‘K’.
    c) Apply Random Forest to achieve classification. Use a suitable method to evaluate the performance of your approach. Tune the parameters of this model- the max_depth and max_features.
    d) Choose a model based on your results from a, b and c. Predict y-hat using predict() with the chosen model, then display the predicted default values using the .value_counts() method.

```python
cd = pd.read_csv('credit_default_dataset.csv')
```

```python
# a) Apply logictic regression to perform classification task. Use suitable method to evaluate the performance of your approach.
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(cd[['LIMIT_BAL']].values, cd['Default'].values)
log_reg_tts = LogisticRegression().fit(x_train, y_train)
log_reg_tts.score(x_test, y_test)
```

```python
# b) Apply K-nearest neighbors to perform classification. Use a suitable method to evaluate the
# performance of your approach. Fine the most suitable value of ‘K’.
from sklearn.neighbors import KNeighborsClassifier as knn

scores = pd.Series()
for i in range(10,1000,100):
     scores.loc[i] = knn(n_neighbors=i).fit(cd[['LIMIT_BAL']].values, 
                                            cd['Default'].values).score(cd[['LIMIT_BAL']].values, cd['Default'].values)
```

```python
scores.max()
```

```python
scores.idxmax()
```

```python
# Train Test Split
x_train, x_test, y_train, y_test = train_test_split(cd[['LIMIT_BAL']].values, cd['Default'].values)
skl_knn_tts = knn(n_neighbors=10).fit(x_train, y_train)
skl_knn_tts.score(x_test, y_test)
```

```python
# c) Apply Random Forest to achieve classification. Use a suitable method to evaluate 
# the performance of your approach. Tune the parameters of this model- the max_depth and max_features.
from sklearn.ensemble import RandomForestClassifier as rf
from sklearn.model_selection import cross_val_score

cross_val_score(rf(n_estimators = 2, max_depth=2, max_features=0.5), cd[['LIMIT_BAL']].values, cd['Default'].values, cv=5).mean()
```

```python
# d) Choose a model based on your results from a, b and c. Predict y-hat using predict() with the chosen model, 
# then display the predicted default values using the .value_counts() method.

x_train, x_test, y_train, y_test = train_test_split(cd[['LIMIT_BAL']].values, cd['Default'].values)
knn_pred =  knn(n_neighbors=10).fit(x_train, y_train)
knn_pred = knn_pred.predict(x_test)
```
